```python
from fuzzywuzzy import process

# Define the list to search through
my_list = ['apple', 'banana', 'orange', 'grape', 'watermelon']

# Define the search string
search_string = 'Mark Cross, Informaton Specialist'

# Use FuzzyWuzzy to find the most similar string in the list
best_match = process.extractOne(search_string, my_list)

# Print the best match
print(f"The best match for '{search_string}' is '{best_match[0]}' with a score of {best_match[1]}")
```

```python
true_string = "Mark Cross Women's Handbags"
string1 = 'Mark Cross, Information Specialist'
string2 = 'Designer Handbags for Women - Womens Designer Fashion'

similarity_score = fuzz.token_set_ratio(true_string, string1)
similarity_score2 = fuzz.token_set_ratio(true_string, string2)

print(f"The token set score between '{true_string}' and '{string1}' \nis {similarity_score}")
print(f"The token set score between '{true_string}' and '{string2}' \nis {similarity_score2}")
```
