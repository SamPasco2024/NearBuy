```python
company_names = [
    'ulla johnson',
    #'mark cross',
    #'le specs',
    #'jl audio',
    #'obey clothing',
    #'roc skin care',
    #'hurley'
]

#searhBar = browser.find_element(By.CLASS_NAME, 'gLFyf')
url_base = 'https://www.google.com/search?q='
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless") # Ensure GUI is off
chrome_options.add_argument("--no-sandbox")

browser = webdriver.Chrome(options=chrome_options) #Chrome() Safari
```

```python
resultsList = []

for zipcode in zipsTop500:
    for company in company_names:
        current_combo = str(company.replace(" ", '+'))+'+'+str(zipcode)
        #print(current_combo)
        
        browser.get(url_base+current_combo) #Get the link
        
        for i in range(0,2):
            browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
            time.sleep(3)
            #print('scrolling...')
            try:
                more = browser.find_element(By.CLASS_NAME, 'RVQdVd')
                more.click()
                #print('load more click!')
            except:
                pass
                #print('pass', current_combo, ' - ', i)
        print("~done scrolling~")
        
        results = browser.find_elements(By.CLASS_NAME, 'yuRUbf')
        print(current_combo + ' - ', str(len(results)))
        for idx, blueLink in enumerate(results, 1):
            resultInfo = {}
            resultInfo['pageRank'] = idx
            resultInfo['zipcode'] = zipcode
            resultInfo['search_combo'] = current_combo
            resultInfo['company'] = company
            resultInfo['link'] = blueLink.find_element(By.TAG_NAME, 'a').get_attribute('href')
            resultInfo['link_name'] = blueLink.find_element(By.TAG_NAME, 'a').find_element(By.CLASS_NAME, 'LC20lb').text
            try:
                resultInfo['link_website'] = blueLink.find_element(By.TAG_NAME, 'a').find_element(By.CLASS_NAME, 'VuuXrf').text
            except:
                pass
            resultsList.append(resultInfo)
        
        browser.quit()
        browser = webdriver.Chrome(options=chrome_options)

        #browser.get('https://www.espn.com/')
        #browser.get('https://www.reddit.com/')
        #browser.get('https://en.wikipedia.org/wiki/Taylor_Swift')
        
        time.sleep(5)
    
```
